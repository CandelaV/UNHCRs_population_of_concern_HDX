# Data preprocessing

```python
import pandas as pd
import os
import csv
```

## Useful Functions

```python
def get_clean_list_of_files(paht):
    file_list = [file for file in os.listdir(paht) if not file.startswith(".")]
    file_list = [file for file in file_list if file.endswith(".csv")]
    return file_list
```

```python
path_out = "../data/processed/neo4j/"
```

## Residing

```python
path_residing_in = "../data/raw/residing/"
file_out_residing = path_out + 'relationsips_residing.csv'
```

```python
residing_list = get_clean_list_of_files(path_residing_in)
residing_list[0:4]
```

```python
column_list = ['#country+residence','#country+origin','#date+year', "#affected+total"]
first_line = [":START_ID(Country)",":END_ID(Country)","year","affected_number"]

countries = set()

with open(file_out_residing, 'a') as f:
    csv.writer(f).writerow(first_line)
    
    for file_in in residing_list:
        
        ds = pd.read_csv(path_residing_in + file_in, skiprows=1)
        ds = ds[column_list]
        #ds['relationship'] = 'ORIGINATE_FROM'
        
        countries.update(ds['#country+residence'].unique())
        countries.update(ds['#country+origin'].unique())
        
        ds.to_csv(f, header=False, index=False)
```

```python
len(countries)
```

```python
with open(path_out + 'countries_residing.csv', 'w') as f:
    f.write("countryId:ID(Country)\n")
    for item in countries:
        f.write('"' + item + '"\n')
```
